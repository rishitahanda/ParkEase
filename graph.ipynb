```python
# Cell 1: Import and Load Data
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

# Load Excel file - change 'xyz.xlsx' to your file path
df = pd.read_excel('xyz.xlsx')
print(f"Columns: {list(df.columns)}")
df.head()

```

```python
# Cell 2: Select X and Y columns and Plot
# Use first two columns as X and Y
x_data = df.iloc[:, 0]  # First column as X
y_data = df.iloc[:, 1]  # Second column as Y


norm = plt.Normalize(min(y_data), max(y_data))
colors = plt.cm.rainbow(norm(y_data))

# Plot bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(x_data, y_data, color=colors)
plt.xlabel("Classification Techniques")
plt.ylabel("Mean Absolute Error")
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Create legend handles manually
legend_handles = [mpatches.Patch(color=colors[i], label=x_data[i]) for i in range(len(x_data))]
plt.legend(handles=legend_handles, loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=len(x_data))

plt.tight_layout()
plt.show()
```
